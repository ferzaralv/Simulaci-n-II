```python
import random as rd
import numpy as np


def dado():
  u=rd.randint(1,6)

  return(u)

def juego():

  pos=0
  cont=0

  while pos<20:

    mov=dado()

    pos+=mov

    match(pos):
      case 3:
        pos=11
      case 13:
        pos=4
      case 15:
        pos=19
      case 17:
        pos=10

    cont+=1
    #print("Turno ", cont, ":", pos)

  return(cont)


def experimento():

  n=1000
  turnos=[]

  for i in range(n):
    turnos.append(juego())

  return(np.mean(turnos))

n=1000
media=[]

for i in range(n):
  media.append(experimento())

print("número de tiradas promedio:",np.mean(media))

```
